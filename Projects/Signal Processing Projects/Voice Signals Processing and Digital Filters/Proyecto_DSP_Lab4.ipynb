#Solución

La solución implementada consiste en analizar grabaciones de voz y determinar la frecuencia fundamental de una vocal específica en un rango de tiempo dado, tanto de voces masculinas como femeninas.

* **Función calculate_frequency:** Esta función calcula la frecuencia
 fundamental de una señal de audio. Toma la señal de audio y la frecuencia de muestreo como entrada, y devuelve la frecuencia fundamental calculada.
* **Función analyze_single_vocal: ** Esta función se encarga de analizar una sola vocal en un archivo de audio. Toma como entrada la ruta del archivo de audio, el rango de muestras que se desea analizar (inicio y fin), y la vocal correspondiente a ese rango. El proceso consiste en:
  * Leer el archivo de audio y graficar la forma de onda completa.
  * Recortar la señal de audio al rango especificado.
  * Graficar la forma de onda de la señal recortada, resaltando el rango seleccionado.
  * Calcular la frecuencia fundamental de la señal recortada utilizando la función calculate_frequency.
  * Mostrar la frecuencia del audio.

* En el programa principal, se especifica el audio que se desea analizar, así como el rango y la vocal correspondiente. Luego, se llama a la función analyze_single_vocal con estos parámetros.


```python
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Función para calcular la frecuencia de la señal de audio
def calculate_frequency(signal, Fs):
    # Retorna la frecuencia de muestreo dividida por la longitud de la señal
    return Fs / len(signal)

# Función para analizar una sola vocal en un archivo de audio
def analyze_single_vocal(file_path, inicio, fin, vocal):
    # Leer el archivo de audio
    Fs, audio = wavfile.read(file_path)

    # Graficar la forma de onda del audio completo
    plt.plot(audio)  # Graficar la señal de audio
    plt.xlabel('Muestras')  # Etiqueta del eje x
    plt.ylabel('Amplitud')  # Etiqueta del eje y
    plt.title('Forma de onda del audio completo')  # Título de la gráfica
    plt.show()  # Mostrar la gráfica

    # Recortar la señal de audio al rango especificado
    signal = audio[inicio:fin]

    # Graficar la forma de onda de la señal recortada
    plt.plot(signal)  # Graficar la señal de audio recortada
    plt.xlabel('Muestras')  # Etiqueta del eje x
    plt.ylabel('Amplitud')  # Etiqueta del eje y
    plt.title(f'Forma de onda en el rango seleccionado (Vocal: {vocal.upper()})')  # Título de la gráfica
    plt.show()  # Mostrar la gráfica

    # Calcular la frecuencia de la señal recortada y mostrarla
    frequency = calculate_frequency(signal, Fs)
    print(f"Vocal: {vocal.upper()}\n")  # Mostrar la vocal
    print(f"\nFrecuencia: {frequency:.2f} Hz")  # Mostrar la frecuencia
```

# Solución

Para resolver el ejercicio, en primer lugar se define el arrange de tiempo que se va a utlizar para generar las señales.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, lfilter, iirfilter, sosfreqz, sosfilt
```

```python
# Tiempo de muestreo y duración
fs = 1000  # Hz
t = np.arange(0, 1, 1/fs)
```

Ahora se definen las tres frecuencias que se van a utilizar para generar la señal, además se definen las tres señales por separado y la señal compuesta por tres señales será la sumatoria de estas tres señales.

```python
# Frecuencias de las señales
f1 = 5  # Hz
f2 = 100 # Hz
f3 = 180 # Hz


# Definición de las señales
Señal1 = np.sin(2 * np.pi * f1 * t)
Señal2 = 1.5* np.sin(2 * np.pi * f2 * t)
Señal3 = 0.5* np.sin(2 * np.pi * f3 * t)

# Señal compuesta por tres frecuencias
SeñalCompuesta = Señal1 + Señal2 + Señal3

# Visualización
plt.figure(figsize=(10, 6))
plt.plot(t, SeñalCompuesta)
plt.xlim(0, 0.5)
plt.title('Señal Compuesta')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

# Recuperación de la señal 1 (5Hz)
A continuación se realizan los filtros FIR e IIR para obtener la señal 1 (señal de 5Hz).

```python
# Filtro FIR
order = 23
fir_filter = firwin(order, cutoff=f1/500, fs=fs)
SeñalRecuperada_fir = lfilter(fir_filter, 1, SeñalCompuesta)

# Filtro IIR
order = 9
iir_filter = iirfilter(N=order, Wn=1.07*f1, btype='low', fs=fs, analog=False, output='sos')
SeñalRecuperada_iir = sosfilt(iir_filter, SeñalCompuesta)

# Visualización de las señales recuperadas
plt.figure(figsize=(12, 6))
plt.plot(t, Señal1, label='Original f1')
plt.plot(t, SeñalRecuperada_fir, label='FIR recuperada f1')
plt.plot(t, SeñalRecuperada_iir, label='IIR recuperada f1')
plt.xlim(0.5, 1)
plt.legend()
plt.title('Comparación de Señales Recuperadas Filtro paso bajo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

# Recuperación de la señal 2 (100Hz)

Ahora se realizan los filtros FIR e IIR para obtener la señal 2 (señal de 100Hz).

```python
# Filtro FIR
order = 39
band = (97/500, 103/500)
fir_filter = firwin(order, band, pass_zero='bandpass')
SeñalRecuperada_fir2 = lfilter(fir_filter, 1, SeñalCompuesta)

# Filtro IIR
order = 3
band2 = (97, 103)
iir_filter = iirfilter(N=order, Wn=band2, btype='bandpass', fs=fs, analog=False, output='sos')
SeñalRecuperada_iir2 = sosfilt(iir_filter, SeñalCompuesta)

# Visualización de las señales recuperadas
plt.figure(figsize=(12, 6))
plt.plot(t, Señal2, label='Original f2')
plt.plot(t, SeñalRecuperada_fir2, label='FIR recuperada f2')
plt.plot(t, SeñalRecuperada_iir2, label='IIR recuperada f2')
plt.xlim(0.9, 1)
plt.legend()
plt.title('Comparación de Señales Recuperadas Filtro paso bajo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

# Recuperación de la señal 3 (180Hz)

Ahora se realizan los filtros FIR e IIR para obtener la señal 3 (señal de 180Hz).

```python
# Filtro FIR
order = 101
fir_filter = firwin(order, 180/(1000/2), pass_zero='highpass')
SeñalRecuperada_fir3 = lfilter(fir_filter, 1, SeñalCompuesta)

# Filtro IIR
order = 8
iir_filter = iirfilter(N=order, Wn=f3, btype='highpass', fs=fs, analog=False, output='sos')
SeñalRecuperada_iir3 = sosfilt(iir_filter, SeñalCompuesta)

# Visualización de las señales recuperadas
plt.figure(figsize=(12, 6))
plt.plot(t, Señal3, label='Original f3')
plt.plot(t, SeñalRecuperada_fir3, label='FIR recuperada f3')
plt.plot(t, SeñalRecuperada_iir3, label='IIR recuperada f3')
plt.xlim(0.9, 1)
plt.legend()
plt.title('Comparación de Señales Recuperadas Filtro paso bajo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

  4. Se utiliza la función *firwin* de scipy, para crear el filtro que se va a utilizar con los parámetros necesarios, en este caso se utilizan los siguientes:

             firwin(orden del filtro, frecuencia de corte, pass_zero='tipo de filtro')
  **Nota:** para la frecuencia de corte es necesario dividir la frecuencia deseada entre la **mitad frecuencia de muestreo**
  5. Ahora se le aplica el filtro a la señal que se desea filtrar, para esto se utiliza la función *lfilter* de scipy y sus parámetros son los siguientes:

            lfilter(filtro_FIR, 1, Señal a filtrar)
